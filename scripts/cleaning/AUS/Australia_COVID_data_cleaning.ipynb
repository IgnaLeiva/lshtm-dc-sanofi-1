```python
#get epi weeks
import pandas as pd
table = pd.read_html('https://www.hpsc.ie/notifiablediseases/resources/epidemiologicalweeks/',header=0)

epi_22 = table[1]
epi_23 = table[2]

epi_22.head()
```

```python
#combine dfs into one table
epi_22_23 = pd.concat([epi_22,epi_23])
epi_22_23

#convert start date to date time
epi_22_23 = epi_22_23[['Year','Week','Start date']]
epi_22_23['Start date'] = pd.to_datetime(epi_22_23['Start date'],format = '%d/%m/%Y')
epi_22_23
```

```python
aus = pd.read_csv('aus_national_covid.csv')
aus.head()
```

```python
aus = aus.rename(columns={'Unnamed: 1':'Year'})
```

```python
dates = []
for i in range(len(aus)):
    date = aus['Date'][i] + '/' + str(aus['Year'][i])
    dates.append(date)

aus['Start date'] = dates
```

```python
aus.head()
```

```python
aus['Start date'] = pd.to_datetime(aus['Start date'],format='%d/%m/%Y')
aus_days = aus[['Start date','Year','AUS']]
aus_days
```

```python
#left join from epi weeks to Australia data
aus_weeks = epi_22_23.merge(aus_days,on='Start date',how='left')
```

```python
#filter out dates after 06/01/2023
aus_weeks = aus_weeks[aus_weeks['Start date'] < '2023-01-07']
aus_weeks = aus_weeks[['Week','Start date','Year_y','AUS']]
aus_weeks = aus_weeks.rename(columns = {'Year_y':'Year','AUS' : 'Count'})
aus_weeks.to_csv('australia_covid_all.csv')
```
